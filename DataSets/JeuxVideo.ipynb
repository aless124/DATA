```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

```python
df = pd.read_csv('./CleanedData2.csv')
df2 = pd.read_csv('./RawData.csv')
df.describe()
df2.describe()

```

```python
sns.set_theme(style="darkgrid")

# Plot the responses for different events and regions
sns.lineplot(data=df,x=df['Year_of_Release'], y="Global_Sales",hue="Genre")
```

```python
sns.set_theme(style="darkgrid")

# Plot the responses for different events and regions
sns.lineplot(data=df,x=df['Year_of_Release'], y="Global_Sales" ,hue="Critic_Score",palette="Spectral", linewidth=1.5)
```

```python
platform = df2.groupby(['Platform']).sum()
number_of_games = df2.groupby(['Platform']).size()
platform.loc[:,"Number_of_Games"] = number_of_games
platform.reset_index()
platforms = platform.reset_index()
platforms = platforms.sort_values(by="Global_Sales", ascending=False)
sns.barplot(data=platforms, x="Platform", y="Global_Sales")
plt.xticks(rotation=90)
```
